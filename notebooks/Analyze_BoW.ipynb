# Import Package

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import locale
from sklearn.feature_extraction.text import CountVectorizer
from plotly.subplots import make_subplots
```

# Load Data

## Load data from analyzed folder

```python
mlbb = pd.read_csv("../data/analyzed/MLBB_Reviews.csv")
hok = pd.read_csv("../data/analyzed/HOK_Reviews.csv")

# Sort dataframes by date column

# Set locale to Indonesia because in column date is using Indonesian Format Date
locale.setlocale(locale.LC_ALL, 'id_ID') 

mlbb['date'] = pd.to_datetime(mlbb['date'], format='%d %B %Y')
hok['date'] = pd.to_datetime(hok['date'], format='%d %B %Y')
mlbb = mlbb.sort_values(by="date")
hok = hok.sort_values(by="date")
print(f"MLBB start from:\t{mlbb.date.min().strftime('%d %B %Y')} - {mlbb.date.max().strftime('%d %B %Y')}")
print(f"HOK start from:\t\t{hok.date.min().strftime('%d %B %Y')} - {hok.date.max().strftime('%d %B %Y')}")
```

```python
fig = make_subplots(rows=1, cols=2, subplot_titles=["Mobile Legends: Bang bang", "Honor of Kings"], specs=[[{'type': 'domain'}, {'type': 'domain'}]])

fig.add_trace(
    go.Pie(labels=mlbb["sentiment"].value_counts().index, values=mlbb["sentiment"].value_counts().values, name="MLBB"),
    row=1, col=1
)
fig.add_trace(
    go.Pie(labels=hok["sentiment"].value_counts().index, values=hok["sentiment"].value_counts().values, name="HoK"),
    row=1, col=2
)

fig.update_layout(
    title=dict(
        text="Sentiment Distribution Comparison (2024 July - 2025 Jan)",
        x=0.5,
        font=dict(size=24)
    )
)
fig.show()
```

Sentimen netral keduanya relatif kecil dengan Mobile Legends 6.59% dan Honor of Kings 8.9%. Secara keseluruhan, Honor of Kings memiliki penerimaan yang sedikit lebih baik dari komunitas pemainnya.

## N-gram analysis

```python
def ngram_analyze(df: pd.DataFrame, text_column: str, n_gram: int = 1, top_n: int = 20) -> None:
    vectorizer = CountVectorizer(ngram_range=(n_gram, n_gram))
    X = vectorizer.fit_transform(df[text_column])

    word_freq = dict(
        zip(
            vectorizer.get_feature_names_out(),
            X.sum(axis=0).tolist()[0]
        )
    )

    wordcloud = WordCloud(width=1280, height=720, background_color="white").generate_from_frequencies(word_freq)
    plt.imshow(wordcloud)
    plt.axis("off")
    plt.show()

    word_freq_df = pd.DataFrame.from_dict(word_freq, orient="index", columns=["frequency"])
    word_freq_df = word_freq_df.sort_values("frequency", ascending=False).head(top_n).reset_index().rename(columns={"index": "word"})

    fig = px.bar(
        word_freq_df,
        x="word",
        y="frequency",
        title=f"Top {top_n} Most Frequent {'Unigrams' if n_gram == 1 else 'Bigrams' if n_gram == 2 else 'Trigrams'}",
        labels={"word": "Word", "frequency": "Frequency"},
    )

    fig.show()
```

### Unigram (1-gram)

### Mobile Legends: Bang bang

```python
ngram_analyze(mlbb, "comment_prep", n_gram=1)
```

### Honor of Kings

```python
ngram_analyze(hok, "comment_prep", n_gram=1)
```

## Bigram (2-gram)

### Mobile Legends: Bang bang

```python
ngram_analyze(mlbb, "comment_prep", n_gram=2)
```

### Honor of Kings

```python
ngram_analyze(hok, "comment_prep", n_gram=2)
```

## Trigram (3-gram)

### Mobile Legends: Bang bang

```python
ngram_analyze(mlbb, "comment_prep", n_gram=3)
```

### Honor of Kings

```python
ngram_analyze(hok, "comment_prep", n_gram=3)
```
